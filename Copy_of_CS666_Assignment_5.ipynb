**Bonus**: Create segmentations (label maps) for the volume using thresholding HU! [33 Points]

```python
# Similar to Window/Level adjustment for visualization, we can threshold
# the volume to highlight the following components using the Hounsfield Units:
# 1) Fat
# 2) Soft Tissue
# 3) Bones
#
# Please create 3 segmentation masks for these structures.
# Then, please visualize each 3 slices per structure to showcase the segmentation.
# Hint: As a reminder, the following code allows thresholding of a numpy array.
#   new_mask = imagevolume.copy()
#   new_mask[new_mask < XXX] = 0
# Hint2: You might need to cast new_mask to int16 not uint16.
```

```python
# TODO: YOUR CODE TO SEGMENT FAT

fat_range = (750, 1050)

f_mask = np.logical_and(final_array >= fat_range[0], final_array <= fat_range[1])
f_mask = f_mask.astype(np.int16)

plt.imshow(f_mask[120, :, :], cmap ='gray')
plt.show
```

```python
# TODO: YOUR CODE TO SEGMENT SOFT TISSUE

tissue_segment_range = (900, 1200)

tissue_segment_mask = np.where(np.logical_and(final_array >= tissue_segment_range[0], final_array <= tissue_segment_range[1]), final_array, 0)
tissue_segment_mask = tissue_segment_mask.astype(np.int16)

plt.imshow(tissue_segment_mask[120, :, :], cmap='gray')
plt.show()
```

```python
# TODO: YOUR CODE TO SEGMENT BONES

bone_range = (1900, 3000)

bone_mask = np.logical_and(final_array >= bone_range[0], final_array <= bone_range[1])
bone_mask = bone_mask.astype(np.int16)

plt.imshow(bone_mask[120, :, :], cmap='gray')
plt.show()
```
